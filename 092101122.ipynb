### Problem 1: Binary Watch (40 points)
 A binary watch has 4 LEDs on the top which represent the hours (0-11), and the 6 LEDs on the bottom represent the minutes (0-59).  
 Each LED represents a zero or one, with the least significant bit  on the right.  

For example, the above binary watch reads "3:25".  
Given a non-negative integer n which represents the number of LEDs that are currently on, return all possible times the watch could represent.

#### The output should be in the following format:  
Return: ["1:00", "2:00", "3:00", "0:01", "0:02", "0:03"]

#### Note:  
The order of output does not matter.

```python
from bt_oo import Backtracking

def exceedBound(v):
    global k
    return sum(v)> k

def mark(v):
    global times, k
    hours = sum(v[i]*(2**i)for i in range(4))
    minutes = sum(v[i]*(2**(i-4))for i in range(4,10))  # Extract the minutes from the last 6 bits
    if hours < 12 and minutes < 60 and sum(v) == k:
        times.append(f"{hours}:{str(minutes).zfill(2)}")
```

```python
# 测试 exceedBound、mark函数
n=10 # 4 bits for hour, 6 bits for minute
x=[0]*n
k=3 # the number of LEDs that are currently on，which will be different in the hidden test
times=[] #the list to store all possible times the watch could represent
binary_watch=Backtracking(n=n,problem_type=1,bound_func=exceedBound,feasible_func=mark)
binary_watch.run()
for time in times:
    print(time)
```
