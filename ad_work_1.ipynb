### Problem 7:  (40 points)
考虑以下数组的 Peak Finding 2D 问题（见Lecture 1）：


下面我们以“二分”法求解。首先完成一些自定义函数：

```python
# 找出二维数组（以列表表示）A 中 mid 列最大元素的位置
def findMax(A,mid):
    max=0
    # YOUR CODE HERE
    for i in range(len(A)):
        if A[i][mid] > A[max][mid]:
            max = i
    return max
```

```python
# 测试 findMax(A,mid)函数
A=[[10,8,10,10],[14,13,12,11],[15,9,11,21],[16,17,19,20]]
assert findMax(A,3) == 2
```

```python
# 判断第 j 列最大值的 A[i,j] 是否为peak
def isPeak(A,i,j):
    flag = True
    # YOUR CODE HERE
    if i > 0 and A[i][j] < A[i - 1][j]:
        flag = False
    if i < len(A) - 1 and A[i][j] < A[i + 1][j]:
        flag = False
    if j > 0 and A[i][j] < A[i][j - 1]:
        flag = False
    if j < len(A[0]) - 1 and A[i][j] < A[i][j + 1]:
        flag = False
    return flag
```

```python
# 测试 isPeak(A,i,j)函数
A=[[10,8,10,10],[14,13,12,11],[15,9,11,21],[16,17,19,20]]
assert isPeak(A,3,1) == False
```

```python
# 在控制台输出二维数组 A 的peak
# 可调用上面的自定义函数
def peakFinding2D(A):
    low = 0
    high = len(A[0])-1
    # YOUR CODE HERE
    while low <= high:
        mid=(low + high) // 2
        max=findMax(A, mid)
        if isPeak(A, max, mid):
            return max, mid
        elif A[max][mid] < A[max][mid - 1]:
            high = mid - 1
        else:
            low = mid + 1
    return -1, -1
```

```python
# 测试 peakFinding2D(A) 函数
import io
from contextlib import redirect_stdout

def test_peakFinding2D():
    A = [[1, 2, 3], [4, 5, 6], [7, 8, 9]]
    f = io.StringIO()
    with redirect_stdout(f):
        peakFinding2D(A)
    output = f.getvalue().strip()
    assert output == "The peak is : A[2,2]:9"

    
```
